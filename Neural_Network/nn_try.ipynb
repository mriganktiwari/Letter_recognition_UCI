```python
def sigmoid (X):
    return 1/(1 + np.exp(-np.float32(X)))

def sigmoidPrime (X):
    a = sigmoid(X)
    return a * (1-a)
```

```python
def forward(X, W1, W2, y):
    Lambda = 0.001
    a1 = X
    z2 = np.dot(X, W1)# + self.b1
    a2 = sigmoid(z2)
    z3 = np.dot(a2, W2)# + self.b2
    
    yHat = sigmoid(z3)
    
    #J = np.sum((y_train * np.log(yHat)) + ((1-y_train) * np.log(1-yHat)))
    J = 0.5*np.sum((y-yHat)**2)/X.shape[0] + (Lambda/2)*(np.sum(W1**2)+np.sum(W2**2))
    return yHat, J
```

```python
def costFunction (y_train, yHat):
    #J = np.sum((y_train * np.log(yHat)) + ((1-y_train) * np.log(1-yHat)))
    #J = 0.5 * np.sum((yHat - y_train)**2)
    J = 0.5*sum((y-self.yHat)**2)/X.shape[0] + (Lambda/2)*(sum(W1**2)+sum(W2**2))
    return J
```

```python
def backPropagation (X, W1, W2, y):
    
    a1 = X
    z2 = np.dot(X, W1)# + self.b1
    a2 = sigmoid(z2)
    z3 = np.dot(a2, W2)# + self.b2
    yHat = sigmoid(z3)
    
    delta3 = np.multiply((yHat - y_train), sigmoidPrime(z3))
    delta2 = np.dot(delta3, W2.T) * sigmoidPrime(z2)
    
    dJdW2 = np.dot(a2.T, delta3)
    dJdW1 = np.dot(a1.T, delta2)
    
    # unrolling Weight matrices into a single vector
    #Dvec = np.append(dJdW1.ravel(),dJdW2.ravel(), axis=0)
    
    return dJdW1, dJdW2#, Dvec
```

```python
# preparing numerical gradient 1

numGrad1 = np.zeros(np.shape(W1))
epsilon_ary1 = np.zeros(np.shape(W1))

for i in range(W1.shape[0]):
    for j in range(W1.shape[1]):
        epsilon_ary1[i,j] = epsilon
        #a1, z2, a2, z3, yHat = forward(X_train, W1-epsilo_ary1, W2)
        cost1 = costFunction(y_train,
                             forward(X_train, W1-epsilon_ary1, W2))
        cost2 = costFunction(y_train,
                             forward(X_train, W1+epsilon_ary1, W2))
        numGrad1[i,j] = (cost2 - cost1) / (2*epsilon)
        epsilon_ary1[i,j] = 0
```

```python
# preparing numerical gradient 2

numGrad2 = np.zeros(np.shape(W2))
epsilon_ary2 = np.zeros(np.shape(W2))

for i in range(W2.shape[0]):
    for j in range(W2.shape[1]):
        epsilon_ary2[i,j] = epsilon
        #a1, z2, a2, z3, yHat = forward(X_train, W1-epsilo_ary1, W2)
        cost1 = costFunction(y_train,
                             forward(X_train, W1, W2-epsilon_ary2))
        cost2 = costFunction(y_train,
                             forward(X_train, W1, W2+epsilon_ary2))
        numGrad2[i,j] = (cost2 - cost1) / (2*epsilon)
        epsilon_ary2[i,j] = 0
```

```python
# gradient checking
## relative error

numGrad = np.hstack((numGrad1.flatten(),
                     numGrad2.flatten()
                    ))
dJdW1, dJdW2 = backPropagation(X_train, W1, W2, y_train)
grad = np.hstack((dJdW1.flatten(),
                  dJdW2.flatten()
                 ))
norm1 = np.linalg.norm(numGrad - grad)
norm2 = np.linalg.norm(numGrad)
norm3 = np.linalg.norm(grad)

relative_error = norm1 / (norm2 + norm3)
```
